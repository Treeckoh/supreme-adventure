```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
class randData():
  def __init__(self,min_t,max_t):
    self.min_t = min_t
    self.max_t = max_t
  def sawtooth_fourier_data(self,num_points, series_length):
    t = np.linspace(self.min_t,self.max_t,num_points)
    x = np.zeros_like(t)
    for i in range(1,series_length+1):
      x = x + (-2/np.pi) * (((-1)**i)/i)*np.sin(2*np.pi*i*t)
    return t,x
  def square_fourier_data(self,num_points, series_length):
    t = np.linspace(self.min_t,self.max_t,num_points)
    x = np.zeros_like(t)
    for i in range(1,series_length+1):
      x = x + (4/np.pi) * ((1)/(2*i-1))*np.sin(2*np.pi*(2*i-1)*t)
    return t,x
  def triangle_fourier_data(self,num_points, series_length):
    t = np.linspace(self.min_t,self.max_t,num_points)
    x = np.zeros_like(t)
    for i in range(1,series_length+1):
      x = x + (-8/(np.pi**2)) * (((-1)**i)/(2*i-1)**2)*np.sin(2*np.pi*(2*i-1)*t)
    return t,x
  def sin_range(self,num_points):
    t = np.linspace(self.min_t,self.max_t,num_points)
    x = np.sin(2*np.pi*t)
    return t,x
```

```python
def main(graph_style = 'scatter'):
  test_cls = randData(min_t = 0, max_t = np.pi)
  point_total = 300
  lenght = 200
  a,b = test_cls.triangle_fourier_data(num_points = point_total, series_length = lenght)
  c,d = test_cls.square_fourier_data(num_points = point_total, series_length = lenght)
  e,f = test_cls.sawtooth_fourier_data(num_points = point_total, series_length = lenght)
  g,h = test_cls.sin_range(num_points = point_total)
  if graph_style == 'plot':
    plt.plot(a,b, linewidth = 3)
    plt.plot(c,d, linewidth = 3)
    plt.plot(e,f, linewidth = 3)
    plt.plot(g,h, linewidth = 3)
  elif graph_style == 'scatter':
    plt.scatter(a,b)
    plt.scatter(c,d)
    plt.scatter(e,f)
    plt.scatter(g,h)
  plt.show()


main('plot')
```
